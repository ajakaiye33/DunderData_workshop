```python
#write a funtion that accepts a singlr parameter to find the number of movies for a givrn rating. Use the function to find the number of movies for ratings: 'R',PG-13 and PG
def mov_rate(rate):
    filt = movy['content_rating']== rate
    return movy[filt]

mov_rate('R')
```
