```python
#import necessary packages
import pandas as pd
```

```python
#read data into pandas
movie = pd.read_csv('data/movie.csv', index_col='title')
movie.head()
```

```python
#select all movies from the 1970
filt1 = movie['year'] >= 1970 
filt2 = movie['year'] < 1980
filt = filt1 & filt2
movie[filt].head()
```

```python
#select movies from the 1970s that had IMDB scores greater than 8
filt3 = movie['imdb_score'] > 8
filty = filt & filt3
movie[filty].head()
```

```python
#select movies that were rated either R, PG-13, or PG
rating_lst = ['R', 'PG-13', 'PG']
filt = movie['content_rating'].isin(rating_lst)
movie[filt].head()
```

```python
#select movies that are either rated Pg-13 or were made after 2010
filt1 = movie['content_rating'] == 'PG-13'
filt2 = movie['year'] > 2010
filt = filt1 | filt2
movie[filt].head()
movie[~filt]
```

```python
# find all the movies that have at least one of the three actors with more than 10,000 facebook likes
filt1 = movie['actor1_fb'] > 10000
filt2 = movie['actor2_fb'] > 10000
filt3 = movie['actor3_fb'] > 10000
filt = filt1 | filt2 | filt3
movie[filt].head(3)
```
